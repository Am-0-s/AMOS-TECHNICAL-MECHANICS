## A simply supported beam $6m$ long is carrying a uniformly distributed load of $5kN/m$ over a length of 3m from the right end. Make a function to draw shear force and bending moment diagrams and calculate maximum bending moment.



# SOLUTION

## Given span $(l)=6m$; Uniformly distributed load $(w)=5kN/m$ and length of the bam $CB$ carrying load $(a)3m$.

## We find reactions $R_{A} and R_{B}$. Taking moments about A and equating the same,

$$ R_{B}\times6 = (5\times3)\times4.5 = 67.5$$

$$R_{B} = \frac{67.5}{6} = 11.25kN$$

## and

$$R_{A} = (5\times3)-11.25= 3.75kN$$

## Shear force
$$F_{A} = +R_{A} = +3.75kN $$

$$F_{C} = +3.75kN $$

$$F_{B} = 3.75-(5\times3) = -11.25kN $$


## Bending Moment
$$ M_{A} = 0 $$

$$ M_{C} = 3.75\times3 = 11.25kNm$$

$$ M_{B} = 0$$

## Shear force and Bending Moment Diagrams


## We know that maximum bending moment will occur at M, where the shear force changes sign. Let x be the distance between C and M. From the geometry of the figure between C and D, we find that

$$ \frac{x}{3.75} = \frac{3-x}{11.25}$$

$$ x=0.75m $$

$$M_{M} = 3.75\times(3+0.75)-5\times\frac{0.75}{2}=12.66kNm$$
