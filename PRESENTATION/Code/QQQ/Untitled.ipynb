# PROBLEM

## Crossing Networks Ltd is a company that deals in bulding bridges. They have approached you as an engineer to design a bridge over River Tana that enables a four-wheeled vehicle of weight *$\color{red}{\text{W}}$* to drive across this brigde. As the Engineer you begin by conceptually idealizing this bridge as a simply supported beam *$\color{red}{\text{AB}}$* of length *$\color{red}{\text{L}}$*.


## Below is the diagram that you come up with.



## The two wheels are a distance *$\color{red}{\text{d}}$* apart and the center of gravity of the vehicle is a distance $\frac{\color{red}{\text{d}}}{3}$ behind the front wheel. You are to sketch shear force and bending moment diagrams and determine the at which the maximum bending moment for further optimal design considerations.



# solution

## Let $R_{R}$ and $R_{F}$ be reaction on the rear wheel and front wheel.

## Taking moment about Front Wheel
 
$$ R_{R} \times d = W \times \frac{d}{3} $$

$$ R_{R} = \frac{W}{3} $$

$$ R_F = W - R_{F} = \frac{2W}{3} $$


## Free body diagram of bridge




## Let $R_{A} and R_{B}$ be reaction force

## Taking about A

$$ (\frac{W}{3}\times a) + (\frac{2w}{3}\times (a+d)) = R_{B} \times L $$

$$ = \frac{W}{3L}(3aL-3a^{2}-3ad+2dL-2ad-2d^{2}) $$

$$ = \frac{W}{3L}(3aL+2dL-5ad-3a^{2}-2d^{2} $$

$$ \frac{d(BM)}{da}= 0 $$

$$ \frac{W}{3L}(3L+0-5d-6a-0) = 0 $$

$$ 3L-5d = 6a $$

$$ a = \frac{3L-5d}{6} $$

$$ (B.M)_{max} = \frac{W}{3L}[3L\frac{(3L-5d)}{6}+2dl-5d\frac{(3L-5d)}{6}-3(\frac{3L-5d}{6})^{2}-2d^{2}] $$

$$= \frac{W}{3L}(\frac{9L^{2}-d^{2}-6Ld}{12}) $$




$$ R_{B} = \frac{W}{3L}[a+2a=2d] $$

$$= \frac{W}{3L}(3a+2d) $$

$$ R_{A} = W - R_{B} $$

$$R_{A}= \frac{W}{3L}(3L-3a-2d) $$




## Shear force calculations

$$ SF_{R_{A}} = R_{A} $$

$$ SF_{\frac{W}{3}} = R_{A}-\frac{W}{3} $$

$$ SF_{\frac{2W}{3}} = R_{A}-\frac{W}{3}-\frac{2W}{3} $$

$$ SF_{R_{B}} = -R_{B} $$


## Bending Moment Calculations

$$ MB_{R_{A}} = 0 $$

$$ MB_{\frac{W}{3}} = R_{A} \times a $$

$$ MB_{\frac{2W}{3}} = R_{A}(a+d) - \frac{W}{3}(d) $$

$$ MB_{R_{B}} = (R_{A}\times L)-\frac{W}{3}(L-a)-\frac{2W}{3}(L-(a+d)) = 0 $$



## Bending moment will be maximum where shear force changes its sign i.e. at front wheel.

## B.M at front wheel:

$$ R_{B} \times (L-a-d) $$

$$= \frac{W}{3L}(3a+2d)(L-a-d) $$

# Global Input Parameters

```python
d=float(input("Enter value for d: "))
W=float(input("Enter value for W: "))
a=float(input("Enter value for a: "))
L=float(input("Enter value for L: "))
q=L-a-d
```

# SHEAR FORCE AND BENDING MOMENT FUNCTION

```python
def sh_bm_calculations():
    #Calcultation for reaction forces
    RB= (W/(3*L))*((3*a)+(2*d))
    RA= (W/(3*L))*((3*L)-((3*a)+(2*d)))
    
    #Shear force calculations
    
    #Shear force at A
    sf_A=RA
    #Shear force at (w/3)
    sf_K=sf_A-(W/3)
    #Shear force at (2W/3)
    sf_KK=sf_K-((2*W)/3)
    #Shear force at RB
    sf_B=-RB
    
    
    #Bending moment calculations
    
    #Bending moment at A
    MB_A = 0
    #Bending moment at (W/3)
    MB_K = RA*a
    #Bending momnet at (2W/3)#
    MB_KK = (RA*(a+d))-((W/3)*d)
    #Bending moment at B
    MB_B=(RA*L)-((W/3)*(L-a))-(((2*W)/3)*(L-(a+d)))
    
    
    #Maximum Bending Moment
    MB_max = RB*(L-a-d)
    
    #printing values
    p1 = f"\nThe result of reaction force at A is {RA:.2f} while at B is {RB:.2f}\n"
    p2 = f"The shear force results at A is {sf_A:.2f}, at w/3 is {sf_K:.2f}, at 2w/3 is {sf_KK:.2f} and at B is {sf_B:.2f}\n"
    p3 = f"The bending moment results at A is {MB_A:.2f}, at w/3 is {MB_K:.2f}, at 2w/3 is {MB_KK:.2f} and at B is {MB_B:.2f}\n"
    p4 = f"The maximum Bending Moment is {MB_max:.2f}\n"
    
    print(p1)
    print(p2)
    print(p3)
    print(p4)
```

# CALLING A FUNCTION

```python
sh_bm_calculations()
```
